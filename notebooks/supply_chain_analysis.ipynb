## ABC Inventory Classification

ABC analysis is used to prioritize inventory
management based on product demand volume.

- A Category: High-priority inventory
- B Category: Medium-priority inventory
- C Category: Low-priority inventory

```python
# Aggregate total orders by product

inventory = (
    sales.groupby('item')['ns_order']
    .sum()
    .reset_index()
)
```

```python
# Classify inventory priority

def classify_inventory(x):

    if x >= 500:
        return 'A'

    elif x >= 200:
        return 'B'

    return 'C'
```

```python
inventory['category'] = (
    inventory['ns_order']
    .apply(classify_inventory)
)
```

```python
# Visualize ABC inventory distribution

plt.figure(figsize=(7,5))

inventory['category'].value_counts().plot(kind='bar')

plt.title('ABC Inventory Classification')
plt.xlabel('Inventory Category')
plt.ylabel('Number of Products')

plt.tight_layout()
plt.show()
```
