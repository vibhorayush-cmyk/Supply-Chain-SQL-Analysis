# Product Demand Analysis

This section identifies the highest-demand products
based on total order volume to support inventory
planning and operational prioritization.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sales = pd.read_csv('sales_orders.csv')

products = pd.read_csv('product_master.csv')

customers = pd.read_csv('customer_master.csv')

suppliers = pd.read_csv('supplier_master.csv')

procurement = pd.read_csv('procurement_orders.csv')
```

## Dataset Overview

The analysis uses operational supply chain datasets
covering customer orders, procurement activity,

## Product Demand Analysis

Analyze high-demand products to support
inventory prioritization and operational planning.

```python
top_products = (
    sales.groupby('product_id')['quantity']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
```

```python
plt.figure(figsize=(10,5))

top_products.plot(kind='bar')

plt.title('Top Products by Order Volume')
plt.xlabel('Product ID')
plt.ylabel('Total Quantity Ordered')

plt.tight_layout()

plt.savefig('top_products.png')

plt.show()
```

## Customer Order Analysis

Identify high-value customers based on
order volume and purchasing activity.

```python
top_customers = (
    sales.groupby('customer_id')['quantity']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
```

```python
plt.figure(figsize=(10,5))

top_customers.plot(kind='bar')

plt.title('Top Customers by Order Volume')
plt.xlabel('Customer ID')
plt.ylabel('Total Quantity Ordered')

plt.tight_layout()

plt.savefig('top_customers.png')

plt.show()
```

## Supplier Performance Analysis

Analyze procurement activity to identify
key suppliers supporting operational demand.

```python
top_suppliers = (
    procurement.groupby('supplier_id')['quantity']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
```

```python
plt.figure(figsize=(10,5))

top_suppliers.plot(kind='bar')

plt.title('Top Suppliers by Procurement Volume')
plt.xlabel('Supplier ID')
plt.ylabel('Total Procurement Quantity')

plt.tight_layout()

plt.savefig('top_suppliers.png')

plt.show()
```

## Operational Order Trend Analysis

Analyze operational demand trends over time
to identify fluctuations and demand spikes.

```python
sales['order_date'] = pd.to_datetime(sales['order_date'])
```

```python
daily_orders = (
    sales.groupby('order_date')['quantity']
    .sum()
)
```

```python
plt.figure(figsize=(12,5))

daily_orders.plot()

plt.title('Operational Order Trends')
plt.xlabel('Date')
plt.ylabel('Order Quantity')

plt.tight_layout()

plt.savefig('order_trends.png')

plt.show()
```

## ABC Inventory Classification

ABC analysis prioritizes inventory
based on operational demand volume.

```python
inventory = (
    sales.groupby('product_id')['quantity']
    .sum()
    .reset_index()
)
```

```python
def classify_inventory(x):

    if x >= 500:
        return 'A'

    elif x >= 200:
        return 'B'

    return 'C'
```

```python
inventory['category'] = (
    inventory['quantity']
    .apply(classify_inventory)
)
```

```python
plt.figure(figsize=(7,5))

inventory['category'].value_counts().plot(kind='bar')

plt.title('ABC Inventory Classification')
plt.xlabel('Inventory Category')
plt.ylabel('Number of Products')

plt.tight_layout()

plt.savefig('abc_inventory.png')

plt.show()
```

## Key Operational Insights

- A small number of products generated a significant share of operational demand.
- Customer demand showed concentration among key accounts.
- Procurement activity highlighted dependency on major suppliers.
- ABC classification identified high-priority inventory categories.
- Order trends revealed operational demand fluctuations over time.
